```python
!pip install -q -U google-generativeai
import google.generativeai as genai

# הגדרת המפתח (שים כאן את המפתח שלך)
genai.configure(api_key="AIzaSyBYVphkLZGMQoZdbSsdeRhU2rg6yfL2Ffk")
model = genai.GenerativeModel('models/gemini-2.5-flash')

# הנחיה ראשונית לסוכן (System Prompt)
context = "אתה מנהל משחק הרפתקאות. השחקן תקוע במבוך. תפקידך לתאר את החדר, להגיב לפעולות שלו ולשלב מדי פעם שאלות ידע כדי להתקדם."

print("--- ברוך הבא למבוך ה-GEM! ---")
chat = model.start_chat(history=[])

# לולאת המשחק
while True:
    user_input = input("מה אתה עושה? (או כתוב 'יציאה'): ")
    if user_input.lower() == 'יציאה':
        break

    # שליחת הפעולה לסוכן וקבלת תגובה
    response = chat.send_message(f"{context} \n השחקן אמר: {user_input}")

    print("\n" + "-"*20)
    print(response.text)
    print("-"*20 + "\n")
```
